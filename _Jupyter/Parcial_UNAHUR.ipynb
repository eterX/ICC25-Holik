## 1 Estados puros

Un estado cuántipo de máxima información puede representarse matemáticamente como un vector de norma uno en un espacio de Hilbert complejo. Es decir, un estado se representa por un vector $|\psi\rangle \in \mathbb{C}^{n}$.

Supongamos que $|\psi\rangle \in \mathbb{C}^{2}$ y que 

$$|\psi\rangle = \sqrt{\frac{2}{3}}|0\rangle +  \sqrt{\frac{1}{3}}|1\rangle$$

Ahora, supongamos que tenemos el estado 

$$|\phi\rangle = \sqrt{\frac{2}{3}}\exp^{\frac{\pi}{3}}|0\rangle +  \sqrt{\frac{1}{3}}\exp^{\frac{\pi}{3}}|1\rangle$$

Calcule las probabilidades de observar los resultados $0$ y $1$, respectivamente, si se realiza una medición en la base computacional, para los estados $|\psi\rangle$ y $|\psi\rangle$. Demuestre que la probabilidad de obtener el resultado cero es la misma para los dos casos (y que lo mismo vale para uno). 

Calcule las probabilidades de observar los resultados ''$+$'' y ''$-$'', respectivamente, si se realiza una medición en la base $\{|+\rangle, |-\rangle\}$, para los estados $|\psi\rangle$ y $|\psi\rangle$. Demuestre que la probabilidad de obtener el resultado $+$ es la misma para los dos casos (y que lo mismo vale para ''$-$''). Notar que la probabilidades de observar ''$+$" y ''$-$" vienen dadas por $|\langle\psi|+\rangle|^{2}$ y $|\langle\psi|-\rangle|^{2}$, respectivamente.

Demuestre que 

$$|\phi\rangle = \exp^{\frac{\pi}{3}}|\psi\rangle $$

Es decir, el estado $|\phi\rangle$ se obtiene a partir de multiplicar a $|\psi\rangle$ por una fase global. Como tales, representan al mismo estado físico. Por ende, sus probabilidades coinciden.


Calcule $$|\psi\rangle\langle\psi|$$ y $$|\phi\rangle\langle\phi|$$

Demuestre que $$|\psi\rangle\langle\psi| = |\phi\rangle\langle\phi|$$

Esta igualdad depende del hecho de que $|\phi\rangle$ se obtiene a partir de multiplicar a $|\psi\rangle$ por una fase global.

Considere ahora el estado:  

$$|\eta\rangle = \sqrt{\frac{2}{3}}|0\rangle +  \sqrt{\frac{1}{3}}\exp^{\frac{\pi}{3}}|1\rangle$$

Calcule las probabilidades $P^{\eta}_{0}$, $P^{\eta}_{1}$, $P^{\eta}_{+}$ y $P^{\eta}_{-}$ (probabilidades asociadas al estado $\eta$). ¿Cómo son estas probabilidades respecto de $P^{\psi}_{0}$, $P^{\psi}_{1}$, $P^{\psi}_{+}$ y $P^{\psi}_{-}$ (probabilidades asociadas al estado $\psi$)?

## 2 Cálculo de autovalores y valores medios de un observable

Supongamos que la energía (el "Hamiltoniano") de un sistema cuántico de dos niveles (un qubit) viene representada por la siguiente matriz Hermítica:

$$ H = -\frac{1}{2} \omega \sigma_{z} -\frac{1}{2} \Omega \sigma_{x}$$

donde 

$$
\begin{align}
\sigma_{z} = \begin{pmatrix}
          1 & 0  \\
          0 & -1 
          \end{pmatrix}
\end{align}$$

y 

$$
\begin{align}
\sigma_{x} = \begin{pmatrix}
          0 & 1  \\
          1 & 0 
          \end{pmatrix}
\end{align}$$

Pruebe que, efectivamente, $H$ es una matriz Hermítica, es decir, $H = \bar{H}^{\dagger}$ (donde la barra denota la conjugada compleja).

Calcule los autovalores de $H$. Recuerte que estos representan los valores posibles de la energía del sistema. Muestre que estos son:

$$\lambda_{1} = \frac{1}{2}\sqrt{\omega^{2}+\Omega^{2}}$$

$$\lambda_{2} = -\frac{1}{2}\sqrt{\omega^{2}+\Omega^{2}}$$

Calcule el valor medio $\langle H \rangle$ de la energía (en función de $\omega$ y $\Omega$) si el estado del sistema es:

$$|\psi\rangle = \frac{1}{\sqrt{2}}(|0\rangle - |1\rangle)$$

Para ello, use las siguientes fórmulas:

$$\langle H \rangle = tr(\rho H)$$

donde $\rho = |\psi\rangle\langle \psi|$

y 
$$\langle H \rangle = \langle\psi|H|\psi\rangle$$

¡Compruebe que de las dos maneras da lo mismo!

# 3 Circuitos cuánticos

Usando Qiskit, escriba un circuito para reproducir el estado:

$$|\psi\rangle = \frac{1}{\sqrt{2}}(|00\rangle + |11\rangle)$$

Luego, calcule el valor medio del observable $\sigma_{z}\otimes\sigma_{z}$ en el estado $|\psi\rangle$, utilizando las siguientes fórmulas:

$$(1) \langle \sigma_{z}\otimes\sigma_{z}\rangle = tr(|\psi\rangle\langle\psi|\sigma_{z}\otimes\sigma_{z})$$

$$(2) \langle \sigma_{z}\otimes\sigma_{z}\rangle = \langle\psi|\sigma_{z}\otimes\sigma_{z}|\psi\rangle$$

Finalmente, corra el circuito en un simulador sin ruido, y obtenga las frecuencias asociadas a los resultados $00$, $01$, $10$ y $11$: $p_{00}$, $p_{01}$, $p_{10}$ y $p_{11}$, respectivamente. Usando los valores obtenidos, calcule: 

$$(3) \langle \sigma_{z}\otimes\sigma_{z}\rangle \approx p_{00}(1) + p_{10}(-1) + p_{01} (-1) + p_{11} (1)$$

Repita el procedimiento asociado a la ecuación (3) usando un simulador ruidoso.
