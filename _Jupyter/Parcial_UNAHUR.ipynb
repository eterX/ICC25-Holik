$$ H = -\frac{1}{2} \omega \sigma_{z} -\frac{1}{2} \Omega \sigma_{x}$$

donde 

$$
\begin{align}
\sigma_{z} = \begin{pmatrix}
          1 & 0  \\
          0 & -1 
          \end{pmatrix}
\end{align}$$

y 

$$
\begin{align}
\sigma_{x} = \begin{pmatrix}
          0 & 1  \\
          1 & 0 
          \end{pmatrix}
\end{align}$$

Pruebe que, efectivamente, $H$ es una matriz Hermítica, es decir, $H = \bar{H}^{\dagger}$ (donde la barra denota la conjugada compleja).
